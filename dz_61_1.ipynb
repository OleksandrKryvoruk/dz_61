## 1. Напишіть код для стиснення зображення з використанням алгоритму JPEG.

```python
import matplotlib.pyplot as plt
from PIL import Image
import os
```

```python
filename = "cat.jpg"
image = Image.open(filename)
print(f'Original size: {os.path.getsize(filename)}')
```

```python
plt.axis('off')
plt.imshow(image)
```

```python
image.save('cat_compr.jpg', format='JPEG', quality=30, optimize=True)
```

```python
print(f'Compressed size: {os.path.getsize("cat_compr.jpg")}')
```
